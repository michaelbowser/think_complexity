```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

G.add_edges_from([
    (1, 2),
    (1, 3),
    (2, 4),
    (3, 4),
    (4, 5)
])
```

```python
plt.figure()
nx.draw(G, with_labels=True)
plt.show()
```

```python
plt.figure()
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=700)
plt.show()
```
