```python
import networkx as nx
G = nx.DiGraph()
```

```python
G.add_node('Alice')
G.add_node('Bob')
G.add_node('Chuck')
```

```python
>>> list(G.nodes())
NodeView(('Alice', 'Bob', 'Chuck'))
```

```python
G.nodes()
```

```python
list(G.nodes())
```

```python
G.add_edge('Alice', 'Bob')
G.add_edge('Alice', 'Chuck')
G.add_edge('Bob', 'Alice')
G.add_edge('Bob', 'Chuck')
```

```python
>>> list(G.edges())
[('Alice', 'Bob'), ('Alice', 'Chuck'),
 ('Bob', 'Alice'), ('Bob', 'Chuck')]
```

```python
COLORS = ['C0', 'C1', 'C2', 'C3']
```

```python
nx.draw_circular(G,
                 node_color=COLORS[0],
                 node_size=2000, 
                 with_labels=True)
```
