```python
import pandas as pd
```

```python
raw_data = [
    {"patient_id": "001", "age": "72", "biomarker": "amyloid", "score": "14.2", "diagnosis": "alzheimers"},
    {"patient_id": "002", "age": "68", "biomarker": "tau", "score": " 9.8 ", "diagnosis": "Alzheimer's"},
    {"patient_id": "003", "age": "unknown", "biomarker": "Amyloid", "score": "15.1", "diagnosis": "control"},
    {"patient_id": "004", "age": "75", "biomarker": "tau", "score": "not available", "diagnosis": "alz"},
    {"patient_id": "005", "age": "70", "biomarker": "amyloid ", "score": "13.7", "diagnosis": "Control"},
    {"patient_id": "006", "age": None, "biomarker": "Tau", "score": "11.4", "diagnosis": "AD"},
]
```

```python
df_raw = pd.DataFrame(raw_data)
```

```python
df = df_raw.copy()
df
```

```python
df.info()
```

```python
df["biomarker"] = df["biomarker"].str.strip().str.lower()
df["diagnosis"] = df["diagnosis"].str.strip().str.lower()
```

```python
diagnosis_map = {
    "alzheimers": "alzheimers",
    "alzheimer's": "alzheimers",
    "alz": "alzheimers",
    "ad": "alzheimers",
    "control": "control"
}

df["diagnosis"] = df["diagnosis"].replace(diagnosis_map)
df
```

```python
df["age"] = pd.to_numeric(df["age"], errors="coerce")
df["score"] = pd.to_numeric(df["score"], errors="coerce")

df
```

```python
df.isna().sum()
```

```python
df_clean = df.dropna(subset=["age", "score"]).copy()
df_clean
```

```python
df_clean["age"] = df_clean["age"].astype(int)
```

```python
df_clean["score"] = df_clean["score"].astype(float)
```

```python
df_clean

```

```python
df_clean.groupby("diagnosis")["score"].mean()
```

```python
df_clean.groupby("biomarker")["score"].mean()
```

```python
df_clean["diagnosis"].value_counts()
```

```python
df_clean.to_csv("alzheimers_cleaned.csv", index=False)
```
